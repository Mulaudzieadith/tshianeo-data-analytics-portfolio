### **Task 2c. Understand the data - Investigate the variables**

In this phase, you will begin to investigate the variables more closely to better understand them.

You know from the project proposal that the ultimate objective is to use machine learning to classify videos as either claims or opinions. A good first step towards understanding the data might therefore be examining the `claim_status` variable. Begin by determining how many videos there are for each different claim status.

```python
# What are the different values for claim status and how many of each are in the data?
data['claim_status'].value_counts()
```

The dataset contains two categories: “claim” and “opinion.” The distribution shows that both categories are present in significant amounts, although one category may have slightly more observations than the other. This indicates that the dataset is reasonably balanced, which is useful for building a classification model.

Next, examine the engagement trends associated with each different claim status.

Start by using Boolean masking to filter the data according to claim status, then calculate the mean and median view counts for each claim status.

```python
# What is the average view count of videos with "claim" status?
data[data['claim_status'] == 'claim']['video_view_count'].mean()
```

```python
# What is the average view count of videos with "opinion" status?
data[data['claim_status'] == 'opinion']['video_view_count'].mean()
```

The average view count differs between claim and opinion videos, suggesting that engagement levels may vary depending on the type of content. Claim-based videos tend to show different viewing patterns compared to opinion-based videos, which may indicate that certain types of content attract more attention from users. This difference could be important when building a classification model, as engagement metrics like view count may help distinguish between claims and opinions.

```python
# Get counts for each group combination of claim status and author ban status
data.groupby(['claim_status', 'author_ban_status']).size()
```

There are noticeable differences in the number of videos across combinations of claim status and author ban status. In particular, claim videos appear more frequently among authors who are under scrutiny or banned compared to opinion videos. This may suggest that content classified as claims is more likely to violate platform policies or spread unverified information, leading to stricter moderation actions. This relationship highlights a potential link between claim-based content and higher moderation risk.

```python
data.groupby('author_ban_status')['video_share_count'].median()
```

```python
# What's the median video share count of each author ban status?
data.groupby('author_ban_status').agg({
    'video_view_count': ['count', 'mean', 'median'],
    'video_like_count': ['mean', 'median'],
    'video_share_count': ['mean', 'median']
})
```

Videos from banned or scrutinized authors show different engagement patterns compared to active authors. In some cases, banned authors may still have high numbers of views, likes, and shares, indicating that their content can reach large audiences despite violating platform policies. This suggests that high engagement does not necessarily mean content is trustworthy or compliant, and it may explain why certain high-performing content is flagged or removed.

```python
# Create engagement ratios
data['likes_per_view'] = data['video_like_count'] / data['video_view_count']
data['comments_per_view'] = data['video_comment_count'] / data['video_view_count']
data['shares_per_view'] = data['video_share_count'] / data['video_view_count']
```

Banned authors show different engagement patterns compared to active authors. In some cases, videos from banned authors still have high numbers of views, likes, and shares, indicating that their content can reach large audiences despite violating platform policies. However, active authors generally show more consistent and stable engagement levels. This suggests that high engagement does not necessarily indicate trustworthy or compliant content, and highly engaging content may still be subject to moderation.

```python
data.groupby(['claim_status', 'author_ban_status']).agg({
    'likes_per_view': ['mean', 'median'],
    'comments_per_view': ['mean', 'median'],
    'shares_per_view': ['mean', 'median']
})
```

Use `groupby()` to compile the information in each of the three newly created columns for each combination of categories of claim status and author ban status, then use `agg()` to calculate the count, the mean, and the median of each group.

```python
data.groupby(['claim_status', 'author_ban_status']).agg({
    'likes_per_view': ['count', 'mean', 'median'],
    'comments_per_view': ['count', 'mean', 'median'],
    'shares_per_view': ['count', 'mean', 'median']
})
```

Claim and opinion videos show differences in engagement patterns. Claim videos tend to have more variable engagement, especially in shares and comments, which suggests they may spread more widely or generate more interaction. Opinion videos may show more stable engagement patterns. Overall, engagement metrics such as likes, comments, and shares appear useful for distinguishing between claim and opinion content.
